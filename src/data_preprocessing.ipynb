```python
#IMPORT ESSENTIAL LIBRARIES FOR DATA MANIPULATION AND NUMERICAL COMPUTATION
import pandas as pd 
import numpy as np
```

```python
path = '../data/nsw-property-sales-data-updated20260127.csv'
#LOAD THE RAW NSW PROPERTY SALES DATASET INTO A PANDAS DATAFRAME
df = pd.read_csv(path)
```

```python
#DISPLAY DATASET STRUCTURE, DATA TYPES, AND NON-NULL COUNTS
df.info()
```

```python
# CALCULATE THE TOTAL NUMBER OF MISSING VALUES FOR EACH COLUMN
df.isnull().sum()
```

```python
#GENERATE SUMMARY STATISTICS FOR ALL NUMERICAL COLUMNS
df.describe()
```

```python
#PREVIEW THE FIRST FIVE ROWS OF THE DATASET
df.head()
```

```python
#REPLACE '?' PLACEHOLDERS WITH NAN AND PREVIEW UPDATED DATAFRAME
df.replace('?',np.nan, inplace = True)
df.head()
```

```python
#CREATE A BOOLEAN MASK TO IDENTIFY ALL MISSING VALUES IN THE DATASET
missing_data = df.isnull()
missing_data.head()
```

```python
#LOOP THROUGH ALL COLUMNS TO COUNT TRUE AND FALSE MISSING VALUES
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("") 
```

```python
#DROP COLUMNS THAT DO NOT CONTRIBUTE TO PRICE PREDICTION
redundant_cols = [
    'Property ID', 'Sale counter', 'Download date / time', 
    'Property name', 'Dealing number'
]
df = df.drop(columns=redundant_cols) 
print(f"Dropped {len(redundant_cols)} redundant columns.")
```

```python
#CALCULATE MISSING PERCENTAGE
missing_pct = (df.isnull().sum() / len(df)) * 100 
print("Missing Value Percentage per Column: ")
print(missing_pct[missing_pct > 0]) 
```

```python
#CALCULATE MEAN AND MEDIAN OF AREA TO DETECT DATA SKEWNESS
avg_area = df["Area"].astype("float").mean(axis=0)
area_median = df['Area'].median()

print(f"Area Mean: {avg_area:.2f}")
print(f"Area Median: {area_median}")
```

```python
#USING GROUPED MEDIAN BY LOCALITY FOR HIGHER ACCURACY
df['Area'] = df['Area'].fillna(df.groupby('Property locality')['Area'].transform('median'))
print(f"Area Median after filling: {df['Area'].median()}")
```

```python
#FILL THE REMAINING WITH GLOBAL MEDIAN  
df['Area'] = df['Area'].fillna(df['Area'].median())
```

```python
#AREA FINAL CHECK
print(f"Missing values in Area: {df['Area'].isnull().sum()}")
```

```python
#AREA TYPE IMPUTATION
area_type_mode = df['Area type'].mode()[0]

df['Area type'] = df['Area type'].fillna(area_type_mode)

print(f"Most frequent Area type (Mode): {area_type_mode}")
print(f"Missing values in Area type: {df['Area type'].isnull().sum()}")
```

```python
#CHECK MISSING RATE OF PURCHASE PRICE 
price_missing_rate = df['Purchase price'].isnull().mean()
print(f'Missing values in Purchase price: {price_missing_rate}')
```

```python
#FILL MISSING CATEGORICAL VALUES USING THE MOST FREQUENT CATEGORY (MODE)
categorical_col = ['Primary purpose'] 
for col in categorical_col:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#USING PLACEHOLDERS FOR HIGH-MISSING OR IDENTIFIER COLUMNS TO AVOID BIAS
df['Zoning'] = df['Zoning'].fillna('Unknown')
df['Property house number'] = df['Property house number'].fillna('N/A')
df['Property unit number'] = df['Property unit number'].fillna('0')
df['Strata lot number'] = df['Strata lot number'].fillna('0')
```

```python
print("---MISSING VALUE CHECK---")
final_missing = df.isnull().sum()
print(final_missing[final_missing >= 0])
```

```python
#DROP REMAINING MINOR MISSING VALUES 
final_cleanup_cols = [
    'Property street name', 'Property locality', 'Property post code',
    'Contract date', 'Settlement date', 'Property legal description'
]

df = df.dropna(subset=final_cleanup_cols)
```

```python
print("---FINAL MISSING VALUE CHECK---")
print(df.isnull().sum())
```

```python
# TASK 4: DATA PREPARATION

# CONVERT CONTRACT DATE TO DATETIME AND EXTRACT YEAR/MONTH FEATURES
df['Contract date'] = pd.to_datetime(df['Contract date'], errors='coerce')

# EXTRACT CONTINUOUS NUMERICAL FEATURES FOR TIME-BASED ANALYSIS
df['Sale_Year'] = df['Contract date'].dt.year
df['Sale_Month'] = df['Contract date'].dt.month

# PREVIEW THE NEWLY CREATED TIME FEATURES
print(df[['Contract date', 'Sale_Year', 'Sale_Month']].head())
```

```python
#CREATE THE NEW FEATURE 
df['Price_per_unit'] = df['Purchase price'] / df['Area']

print(df[['Purchase price', 'Area', 'Price_per_unit']].head())

#CHECK STATISTICS TO FIND POTENTIAL OUTLIERS 
print("Summary Statistics for Price per Unit: ")
print(df['Price_per_unit'].describe())
```

```python
# CLEANING PRICE PER UNIT (Removing Inf and Outliers)

#REPLACE INFINITY VALUES WITH NAN
df['Price_per_unit'] = df['Price_per_unit'].replace([np.inf, -np.inf], np.nan)

#DROP ROWS WHERE Price_per_unit IS NAN 
df = df.dropna(subset=['Price_per_unit'])

#FILTER OUTLIERS
#KEEPING TRANSACTIONS WITH REALISTIC PRICES($10 TO $500,000 PER M2 ) 
df = df[(df['Price_per_unit'] > 10) & (df['Price_per_unit'] < 500000)]

#FINAL VERIFICATION
print("---CLEANED SUMMARY STATISTICS---")
print(df['Price_per_unit'].describe())
```

```python
print(f"Final dataset shape: {df.shape}")
```

```python
# TASK 5: DATA PREPARATION - FEATURE SCALING & LEAKAGE-SAFE PREPROCESSING

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

# 1. DEFINE FEATURES AND TARGET VARIABLE
# WE SELECT AREA AND TEMPORAL FEATURES TO PREDICT PRICE PER UNIT AREA
numeric_features = ['Area', 'Sale_Year', 'Sale_Month']
X = df[numeric_features]
y = df['Price_per_unit']

# 2. IMPLEMENT TIME-BASED SPLIT (FOR RESEARCH QUESTION 3 REALISM)
# SHUFFLE IS SET TO FALSE TO PRESERVE TEMPORAL ORDER (PREDICTING FUTURE FROM PAST)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 3. INITIALIZE PREPROCESSING PIPELINE
# STANDARDSCALER ENSURES ALL FEATURES HAVE MEAN=0 AND STD=1 FOR LINEAR/SVR MODELS
pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# 4. EXECUTE LEAKAGE-SAFE TRANSFORMATION
# IMPORTANT: FIT ONLY ON TRAINING DATA TO PREVENT LOOK-AHEAD BIAS (DATA LEAKAGE)
X_train_scaled = pipeline.fit_transform(X_train)
X_test_scaled = pipeline.transform(X_test)

print("SUCCESS: FEATURES STANDARDIZED AND LEAKAGE-SAFE PROTOCOL APPLIED!")
```
