# Phân tích dữ giá nhà tại Melbourne (Melbourne housing price Data Analysis)

## 1. Giới thiệu bài toán

Giá bất động sản có xu hướng tăng nhanh hay chậm phụ thuộc vào nhiều yếu tố. Việc phân tích và làm rõ mối quan hệ giữa giá cả bất động sản và những đặc trưng như số phòng, diện tích, khu vực, etc. giúp các cá nhân lập kế hoạch và lựa chọn nhà ở một cách hợp lý, phù hợp hơn. Dựa vào dữ liệu thu thập được, ta sẽ phân tích làm rõ:
- Mức ảnh hưởng của các tính chất của bất động sản (diện tích, số lượng phòng, khoảng cách đến khu trung tâm thương mại,...) đến giá trị của nó.
- Dự đoán xu hướng của giá nhà dựa vào các tính chất của bất động sản.

## 2. Mục tiêu phân tích

Dự án này tập trung vào việc phân tích dữ liệu giá cả nhà đất tại Melbourne và các đặc trưng của nó (như diện tích, số lượng phòng các loại, diện tích, etc.) để trả lời các câu hỏi sau:
- Mối quan hệ giữa các tính chất của một căn bất động sản và giá thành là gì?
- Các tính chất như vị trí, diện tích, số lượng phòng, ect. có ảnh hưởng như thế nào đến giá cả?
- Có thể dự đoán sự thay đổi giá trị bất động sản dựa vào những tính chất của nó hay không?

## 3. Các yếu tố ảnh hưởng đến giá cả BĐS

Dựa trên dữ liệu thu thập, có một số yếu tố quan trọng có thể ảnh hưởng đến giá BĐS (có thể thay đổi trong quá trình xử lý, phân tích):

- **Suburb**: Vị trí của khu bất động sản (e.g Canterbury, Brighton, Middle Park, etc.)
- **Rooms**: Tổng số phòng
- **Type**: Loại hình BĐS (h = house, cottage, villa, etc; u = unit, duplex, townhouse; t = townhouse/terrac)
- **Price**: Giá cả của BĐS (Dollar)
- **Date**: Ngày giao dịch (Từ tháng 1-2016 đến tháng 9-2017)
- **Distance**: Khoảng cách đến khu trung tâm thương mại
- **Bedroom2**: Số lượng phòng ngủ
- **Bathroom**: số lượng phong tắm
- **Car**: Số bãi đỗ xe
- **Landsize**: Tổng diện tích BĐS
- **Lattitude**: Vĩ độ
- **Longtitude**: Kinh độ
- **Regionname**: Vị trí theo vùng (e.g Southern Metropolitan)

## 4. Cách tiếp cận phân tích

Để đạt được mục tiêu trên, dự án sẽ thực hiện các bước sau:
1. **Thu thập và làm sạch dữ liệu**: Giữ lại những cột cần thiết, loại bỏ dữ liệu bị thiếu, trùng lặp, đưa kiểu dữ liệu về loại phù hợp, và lọc dữ liệu từ các nguồn tin cậy.
2. **Phân tích dữ liệu bằng SQL**: Truy vấn dữ liệu để tìm hiểu xu hướng giá BĐS theo các đặc trưng của nó.
3. **Phân cụm dữ liệu (KMeans) và PCA**: Xác định các nhóm BĐS có đặc điểm tương đồng về giá cả và các tính chất phù hợp (ta chọn cách tính chất mà giá cả khi chia thành các nhóm có sự khác biệt rõ rệt).
4. **Trực quan hóa dữ liệu**: Sử dụng biểu đồ để minh họa mối quan hệ giữa các yếu tố tác động.
5. **Xây dựng mô hình dự đoán**: Sử dụng hồi quy hoặc mô hình học máy để dự đoán sức giá cả dựa trên dữ liệu khác của BĐS.
6. **Tạo dashboard với Power BI/Tableau**: Cung cấp các báo cáo trực quan giúp người dùng có cái nhìn tổng quan về mối quan hệ giữa giá cả và các đặc trưng của BĐS.

## 5. Kỳ vọng kết quả

- Hiểu rõ hơn về tác động của các đặc trưng BĐS với giá cả.
- Phân loại và hiểu rõ các nhóm đối tượng dựa trên các tính chất phù hợp của BĐS và giá cả của nó.
- Xây dựng mô hình dự đoán giúp các cá nhân cân nhắc về giá cả, từ đó đưa ra quyết định mua bán phù hợp hơn.
